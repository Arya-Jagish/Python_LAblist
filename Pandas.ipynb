```python
import pandas as pd

# Create two Pandas Series
series1 = pd.Series([1, 2, 3, 4, 5])
series2 = pd.Series([5, 4, 3, 2, 1])

# Perform addition
addition_result = series1 + series2

# Perform subtraction
subtraction_result = series1 - series2

# Perform multiplication
multiplication_result = series1 * series2

# Perform division (watch out for division by zero)
division_result = series1 / series2

# Create a DataFrame to display the results
results_df = pd.DataFrame({
    'Series1': series1,
    'Series2': series2,
    'Addition': addition_result,
    'Subtraction': subtraction_result,
    'Multiplication': multiplication_result,
    'Division': division_result
})

# Display the results
print(results_df)
```

```python
import pandas as pd

# Create two sample Series
series1 = pd.Series([1, 2, 3, 4, 5])
series2 = pd.Series([3, 4, 5, 6, 7])

# Find items in series1 that are not in series2
not_common_in_series1 = series1[~series1.isin(series2)]

# Find items in series2 that are not in series1
not_common_in_series2 = series2[~series2.isin(series1)]

# Concatenate the two resulting Series to get items not common in either series
not_common = pd.concat([not_common_in_series1, not_common_in_series2])

# Display the items not common in both series
print(not_common)
```

```python
import pandas as pd

# Sample Series
data = pd.Series(['apple', 'banana', 'cherry', 'dog', 'elephant', 'fig', 'grape'])

# Function to count vowels in a word
def count_vowels(word):
    vowels = 'aeiouAEIOU'
    count = 0
    for char in word:
        if char in vowels:
            count += 1
    return count

# Filter words with at least two vowels
filtered_words = data[data.apply(lambda x: count_vowels(x) >= 2)]

print(filtered_words)
```

```python
import pandas as pd

# Create a sample DataFrame
data = {'Age': [25, 30, 35, 40, 45, 22, 28, 32, 38, 43],
        'Salary': [50000, 60000, 70000, 80000, 90000, 45000, 55000, 65000, 75000, 85000]}

df = pd.DataFrame(data)

# Define bins and labels for the 'Age' column
age_bins = [20, 30, 40, 50]
age_labels = ['20-29', '30-39', '40-49']

# Convert 'Age' to categorical
df['Age_Category'] = pd.cut(df['Age'], bins=age_bins, labels=age_labels, right=False)

print("Original DataFrame:")
print(df)

# Define bins and labels for the 'Salary' column
salary_bins = [40000, 60000, 80000, 100000]
salary_labels = ['Low', 'Medium', 'High']

# Convert 'Salary' to categorical
df['Salary_Category'] = pd.cut(df['Salary'], bins=salary_bins, labels=salary_labels, right=False)

print("\nDataFrame with Continuous Columns Converted to Categorical:")
print(df)
```

```python
import pandas as pd

# Define the first two DataFrames
data1 = {'student_id': ['S1', 'S2', 'S3', 'S4', 'S5', 'S7', 'S8', 'S9', 'S10', 'S11', 'S12', 'S13'],
         'exam_id': [23, 45, 12, 67, 21, 55, 33, 14, 56, 83, 88, 12]}

exam_data = pd.DataFrame(data1)

# Define the third DataFrame to be merged
data2 = {'student_id': ['S1', 'S6', 'S7', 'S10', 'S13'],
         'exam_score': [95, 88, 78, 92, 87]}

exam_scores = pd.DataFrame(data2)

# Join the first two DataFrames along rows
merged_data = pd.concat([exam_data, exam_scores], ignore_index=True)

# Print the merged DataFrame
print("Merged DataFrame (first two DataFrames joined along rows):")
print(merged_data)

# Merge the merged_data DataFrame with the third DataFrame along the 'student_id' column
final_data = merged_data.merge(exam_scores, on='student_id', how='left')

# Print the final merged DataFrame
print("\nFinal Merged DataFrame (merged with the third DataFrame):")
print(final_data)
```

```python
import pandas as pd

# Create the DataFrame
data = {'school': ['s001', 's002', 's003', 's001', 's002', 's004'],
        'class': ['V', 'V', 'VI', 'VI', 'V', 'VI'],
        'name': ['Alberto Franco', 'Gino Mcneill', 'Ryan Parkes', 'Eesha Hinton', 'Gino Mcneill', 'David Parkes'],
        'date_Of_Birth': ['15/05/2002', '17/05/2002', '16/02/1999', '25/09/1998', '11/05/2002', '15/09/1997'],
        'age': [12, 12, 13, 13, 14, 12],
        'height': [173, 192, 186, 167, 151, 159],
        'weight': [35, 32, 33, 30, 31, 32],
        'address': ['street1', 'street2', 'street3', 'street1', 'street2', 'street4']}

df = pd.DataFrame(data)

# Split the DataFrame by school code and calculate mean, min, and max age for each school
result = df.groupby('school')['age'].agg(['mean', 'min', 'max'])

print("Mean, Min, and Max Age for Each School:")
print(result)
```

```python
import pandas as pd

# Create the DataFrame
data = {'Id': [1, 2, 1, 1, 2, 1, 2],
        'type': [10, 15, 11, 20, 21, 12, 14],
        'book': ['Math', 'English', 'Physics', 'Math', 'English', 'Physics', 'English']}

df = pd.DataFrame(data)

# Split the DataFrame into groups based on all columns and calculate value counts
grouped = df.groupby(['Id', 'type', 'book']).size().reset_index(name='count')

print("Value Counts for Each Group:")
print(grouped)
```
