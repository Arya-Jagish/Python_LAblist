```python
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn.cluster import KMeans

# Load the data
data = pd.read_csv("Mall_Customers.csv")

# Select features for clustering
X = data[['Age', 'Annual Income (k$)', 'Spending Score (1-100)']]

# Perform K-Means clustering
num_clusters = 5
kmeans = KMeans(n_clusters=num_clusters)
kmeans.fit(X)

# Add cluster labels to the data
data['Cluster'] = kmeans.labels_

# Create a 3D scatter plot
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

colors = ['red', 'green', 'blue', 'cyan', 'magenta']

for cluster in range(num_clusters):
    clustered_data = data[data['Cluster'] == cluster]
    ax.scatter(clustered_data['Age'], clustered_data['Annual Income (k$)'], clustered_data['Spending Score (1-100)'],
               c=colors[cluster], label=f'Cluster {cluster}')

# Set labels for the axes
ax.set_xlabel('Annual Income (k$)')
ax.set_ylabel('Spending Score (1-100)')
ax.set_zlabel('Age')

# Show the legend
ax.legend()

plt.title('3D Cluster Visualization')
plt.show()
```

```python
import pandas as pd
import numpy as np
from mlxtend.frequent_patterns import association_rules,apriori
```

```python
data=pd.read_excel('Online Retail.xlsx')
data.head()
```

```python
data.Country.unique()
```

```python
data['Description']=data['Description'].str.strip()
data.dropna(axis=0,subset=['InvoiceNo'],inplace=True)
data['InvoiceNo']=data['InvoiceNo'].astype('str')
data=data[~data['InvoiceNo'].str.contains('C')]
```

```python
basket_France=(data[data['Country']=='France'].groupby(['InvoiceNo','Description'])['Quantity'].sum().unstack().reset_index().fillna(0).set_index('InvoiceNo'))
```

```python
def hot_encode(X):
    if X<=0:
        return 0
    if X>=1:
        return 1
    
basket_encode=basket_France.applymap(hot_encode)
basket_France=basket_encode
```

```python
frquent_items=apriori(basket_France,min_support=0.05,use_colnames=True)
rules=association_rules(frquent_items,metric="lift",min_threshold=1)
rules=rules.sort_values(['confidence','lift'],ascending=[False,False])
rules.head()
```
